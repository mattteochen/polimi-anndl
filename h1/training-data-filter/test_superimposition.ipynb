```python
import numpy as np
import matplotlib.pyplot as plt

path_data = 'training_set.npz'
```

```python
data = np.load(path_data)
X = data['images']
Y = data['labels']
```

```python
# Plot 30 random images in a 5x6 grid

fig, ax = plt.subplots(5, 6, figsize=(12, 10))
for i in range(5):
    for j in range(6):
        idx = np.random.randint(0, X.shape[0])
        ax[i, j].imshow(X[idx])
        ax[i, j].set_title(idx)
        ax[i, j].axis('off')
plt.show()
```

```python
# sets to prevent duplicates
shrek = set([13290,12442,12635,12849,13016,11981,12873,12832,13454,13060,12419,13455,13464,12379,12987,12740,12439,12029,12148,13533,12539])
rick_astley = set([13688,13690,13679,13620,13629,13582,13681,13653,13721])
```

```python
# Plot all 21 shrek images in a 3x7 grid
fig, ax = plt.subplots(3, 7, figsize=(12, 6))
for i, idx in enumerate(shrek):
    ax[i//7, i%7].imshow(X[idx])
    ax[i//7, i%7].set_title(idx)
    ax[i//7, i%7].axis('off')
plt.show()

```

```python
# Check that all shrek images are identical
shrek_images = X[list(shrek)]
shrek_labels = Y[list(shrek)]
print(np.all(shrek_images == shrek_images[0]))
print(np.all(shrek_labels == shrek_labels[0]))

shrek_labels[0:5].tolist()
```

```python
# Count shrek images over the whole dataset
shrek_count = np.sum(np.all(X == shrek_images[0], axis=(1, 2)))
print(shrek_count)
```

```python
# Plot all 9 rick astley images in a 3x3 grid
fig, ax = plt.subplots(3, 3, figsize=(6, 6))
for i, idx in enumerate(rick_astley):
    ax[i//3, i%3].imshow(X[idx])
    ax[i//3, i%3].set_title(idx)
    ax[i//3, i%3].axis('off')

plt.show()
```

```python
# Check that all rick astley images are identical
rick_astley_images = X[list(rick_astley)]
rick_astley_labels = Y[list(rick_astley)]

print(np.all(rick_astley_images == rick_astley_images[0]))
print(np.all(rick_astley_labels == rick_astley_labels[0])) # Apparently all rick astley images are identical in label too but...

rick_astley_labels[0:5].tolist()
```

```python
# Count rick astley images over the whole dataset
rick_astley_count = np.sum(np.all(X == rick_astley_images[0], axis=(1, 2)))
print(rick_astley_count)
```

```python
# Print 49 images in a 7x7 grid, all random of label 5

fig, ax = plt.subplots(7, 7, figsize=(12, 12))
for i in range(7):
    for j in range(7):
        idx = np.random.choice(np.where(Y == 5)[0])
        ax[i, j].imshow(X[idx])
        ax[i, j].set_title(idx)
        ax[i, j].axis('off')

plt.show() #... but they dont't match visually with the label 5 images. So we can toss them out as well. 
```
